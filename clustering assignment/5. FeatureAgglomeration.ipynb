```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import FeatureAgglomeration
agglo=FeatureAgglomeration(n_clusters=2,  metric='euclidean',linkage='ward')
transformed_x=agglo.fit_transform(x)
```

```python
from sklearn.cluster import AgglomerativeClustering
agglo=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
label=agglo.fit_predict(x)
label
```

```python
supervised=dataset
```

```python
supervised['cluster_groups']=label
```
