```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import HDBSCAN
hdbscan=HDBSCAN(min_cluster_size=5, min_samples=None,algorithm='auto')
y=hdbscan.fit_predict(x)
y
```

```python
supervised=dataset
```

```python
supervised['cluster_groups']=y
```

```python
supervised
```

```python
dir(hdbscan)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],
                 hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
