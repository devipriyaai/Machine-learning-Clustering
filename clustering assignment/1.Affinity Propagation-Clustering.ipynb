```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import AffinityPropagation
list1=[]
for i in range(1,11):
    aff_propagation=AffinityPropagation(damping=0.5,max_iter=200)
    aff_propagation.fit(x)
    n_clusters = len(set(aff_propagation.labels_))
    list1.append(n_clusters)
```

```python
damping_values=np.linspace(0.5,1.0,10) # Damping values in the valid range
```

```python
plt.plot(damping_values,list1, marker='o')
plt.title('Number of Clusters vs. Damping in Affinity Propagation')
plt.xlabel('Damping Value')
plt.ylabel('Number of Clusters')
plt.show()
```

```python
aff_propagation= AffinityPropagation(damping=0.5, max_iter=200)
y_affinity= aff_propagation.fit_predict(x)
y_affinity
```

```python
supervised=dataset
```

```python
supervised['cluster_groups']=y_affinity
```

```python
supervised
```

```python
centroids=aff_propagation.cluster_centers_
```

```python
centroids
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,
             legend=True,legend_out=True)
```
