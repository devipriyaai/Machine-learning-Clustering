```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import MeanShift
clustering=MeanShift(bandwidth=10,bin_seeding=False,cluster_all=True)
label=clustering.fit_predict(x)
label
```

```python
supervised=dataset
```

```python
supervised['cluster_groups']=label
```

```python
supervised
```

```python
dir(clustering)
```

```python
centroids=clustering.cluster_centers_
```

```python
centroids
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],
                 hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
